# DML

```python
data = [(5,100,'aa','1'),(6,200,'bb','1'),(7,300,'cc','1'),(7,400,'dd','1')]

df = spark.createDataFrame(data,['cust_id','income','name','tip'])

df.write.format('delta')\
    .mode('append')\
    .save('/Volumes/deltalakesharnam/default/deltavol/dmlsink/')
```

```python
%sql
SELECT * FROM delta.`/Volumes/deltalakesharnam/default/deltavol/dmlsink/`
```

# UPDATE

```python
%sql
UPDATE delta.`/Volumes/deltalakesharnam/default/deltavol/dmlsink/`
SET income = 1000
WHERE cust_id = 5
```

```python
%sql
SELECT * FROM delta.`/Volumes/deltalakesharnam/default/deltavol/dmlsink/`
ORDER BY cust_id
```

# UPSERT

```python
data = [(1,100,'xyz',1),(9,200,'bb',1),(10,300,'cc',1)]

df = spark.createDataFrame(data,['cust_id','income','name','tip'])

display(df)
# df.write.format('delta')\
#     .mode('append')\
#     .save('/Volumes/deltalakesharnam/default/deltavol/dmlsink/')
```

```python
from delta.tables import DeltaTable
```

```python
dlt_obj = DeltaTable.forPath(spark,"/Volumes/deltalakesharnam/default/deltavol/dmlsink/")

dlt_obj.alias('trg').merge(df.alias('src'),"trg.cust_id = src.cust_id")\
        .whenMatchedUpdateAll()\
        .whenNotMatchedInsertAll()\
        .execute()
```
