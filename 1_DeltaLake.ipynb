# CREATE DELTA TABLE

### USING SQL API

```python
# %sql
# DROP TABLE deltalakesharnam.default.first_table
```

```python
%sql
CREATE TABLE deltalakesharnam.default.first_table
(
    id INT NOT NULL ,
    salary int
)
-- USING DELTA(no need as by default it creates delta table)
```

### USING DELTA API

```python
spark.sql("SELECT current_catalog()").show()
```

```python
%sql
USE CATALOG deltalakesharnam
```

```python
from delta.tables import DeltaTable
```

```python
DeltaTable.createIfNotExists(spark)\
    .tableName('default.firstdeltaapi')\
    .addColumn('id',"INT")\
    .addColumn('salary',"INT")\
    .execute()
```

# GENRATED COLUMNS

### IDENTITY COLUMNS

```python
from pyspark.sql.types import *
from delta.tables import IdentityGenerator
```

```python
DeltaTable.create(spark)\
    .tableName("deltalakesharnam.default.identity_table")\
    .addColumn("id_col1",dataType=LongType(),generatedAlwaysAs=IdentityGenerator())\
    .addColumn("salary",dataType=IntegerType())\
    .addColumn("name",dataType=StringType())\
    .execute()
```

### COMPUTED COLUMNS

```python
DeltaTable.create(spark)\
    .tableName("deltalakesharnam.default.identity_table")\
    .addColumn("id_col1",dataType=LongType(),generatedAlwaysAs=IdentityGenerator())\
    .addColumn("salary",dataType=IntegerType())\
    .addColumn("name",dataType=StringType())\
    .execute()
```
