# SCHEMA ENFORCEMENT

```python
data = [(1,100,'aa','1'),(2,200,'bb','1'),(3,300,'cc','1'),(4,400,'dd','1')]

df = spark.createDataFrame(data,['id','salary','name','test_col'])

display(df)
```

```python
df.write.format('delta')\
    .mode('append')\
    .save('/Volumes/deltalakesharnam/default/deltavol/writesink/')
```

# SCHEMA EVOLUTION

```python
df.write.format('delta')\
    .mode('append')\
    .option('mergeSchema',True)\
    .save('/Volumes/deltalakesharnam/default/deltavol/writesink/')
```

SCHEMA OVERWRITE

```python
data = [(1,100,'aa','1'),(2,200,'bb','1'),(3,300,'cc','1'),(4,400,'dd','1')]

df = spark.createDataFrame(data,['cust_id','income','name','tip'])

display(df)
```

```python
df.write.format('delta')\
    .mode('overwrite')\
    .save('/Volumes/deltalakesharnam/default/deltavol/writesink/')
```

```python
df.write.format('delta')\
    .mode('overwrite')\
    .option('overwriteSchema',True)\
    .save('/Volumes/deltalakesharnam/default/deltavol/writesink/')
```
