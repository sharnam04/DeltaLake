```python
data = [(1,100,'xyz',1),(9,200,'bb',1),(10,300,'cc',1)]

df = spark.createDataFrame(data,['cust_id','income','name','tip'])

display(df)
```

```python
df.write.format('delta')\
        .mode('append')\
        .save('/Volumes/deltalakesharnam/default/deltavol/schemalevel/')
```

```python
from delta.tables import DeltaTable

# Enable column mapping on the Delta table
spark.sql("""
    ALTER TABLE delta.`/Volumes/deltalakesharnam/default/deltavol/schemalevel/`
    SET TBLPROPERTIES (
        'delta.columnMapping.mode' = 'name',
        'delta.minReaderVersion' = '2',
        'delta.minWriterVersion' = '5'
    )
""")

# Now rename the column
spark.sql("""
    ALTER TABLE delta.`/Volumes/deltalakesharnam/default/deltavol/schemalevel/`
    RENAME COLUMN name TO cust_name
""")
```
