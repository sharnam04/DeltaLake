# OPTIMIZATION 

```python
df = spark.read.table('deltalakesharnam.default.clonetbl')
```

```python
display(df)
```

```python
## we ran this 3 times
df.write.format('delta')\
    .mode('append')\
    .save('/Volumes/deltalakesharnam/default/deltavol/optimization/')
```

```python
%sql
OPTIMIZE delta.`/Volumes/deltalakesharnam/default/deltavol/optimization`
```
